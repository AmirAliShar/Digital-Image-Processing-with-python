```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=plt.imread("rose.jpeg")
# Show with Matplotlib (needs conversion BGR -> RGB)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title("Image in RGB")
plt.axis("off")
plt.show()
```

```python

plt.figure(figsize=(12,16))
img = cv2.imread("rose.jpeg")

plt.subplot(1,2,1)
plt.title("Without Conversion")
plt.imshow(img)  # colors wrong because Because Matplotlib is designed to follow standard image display convention (RGB)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("With BGRRGB")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # correct
plt.axis("off")

plt.show()
```

# 4. Converting Color Models

```python
# Load the image
img=cv2.imread("rose.jpeg")
plt.figure(figsize=(16,16))

#BGR to RGB
RGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

# Grayscale
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#Grayscale  RGB
G2RGB=cv2.cvtColor(gray,cv2.COLOR_GRAY2RGB)

# HSV (Hue, Saturation, Value)
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

plt.subplot(1,5,1); plt.imshow(img); plt.title("BGR"); plt.axis(False)

plt.subplot(1,5,2); plt.imshow(RGB); plt.title("BGR2RGB"); plt.axis(False)

plt.subplot(1,5,3);plt.imshow(gray);plt.title("GBR2Gray");plt.axis(False)

plt.subplot(1,5,4);plt.imshow(G2RGB);plt.title("Gray2RGB");plt.axis(False)


plt.subplot(1,5,5);plt.imshow(hsv);plt.title("GBR2HSV");plt.axis(False);plt.show()

```
