```python
import cv2
import matplotlib.pyplot as plt
```

```python
#loading the image
img=cv2.imread("rose.jpeg")
RGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(RGB)
plt.show()
```

```python
# (rows, cols, channels)
print("shape ",img.shape)

# total number of pixels
print("Size ",img.size)

print("Datatype:", img.dtype)
```

# Cropping & Resizing

```python
# Cropping
crop = img[50:200, 100:300]

# Resizing
resize = cv2.resize(img, (200, 200))   # fixed size
resize_scale = cv2.resize(img, None, fx=0.5, fy=0.5)  # half size

plt.figure(figsize=(10,10))
plt.subplot(1,3,1); plt.imshow(RGB); plt.title("Original"); plt.axis("off")
plt.subplot(1,3,3); plt.imshow(cv2.cvtColor(resize, cv2.COLOR_BGR2RGB)); plt.title("Resized"); plt.axis("off")
plt.subplot(1,3,2); plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)); plt.title("Cropped"); plt.axis("off")

plt.show()
```

```python
# save the image RGB image
cv2.imwrite("RGB.jpeg",RGB)
```
